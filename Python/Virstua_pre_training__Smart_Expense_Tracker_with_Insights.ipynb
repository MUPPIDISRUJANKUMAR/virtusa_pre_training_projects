```python
from datetime import date
import csv
import uuid

def add_expenses():
    expenses_id=uuid.uuid4().hex[:8]
    choice=input("Use today's date? (Y/N)")
    if choice.upper()=="Y":
        expenses_date=date.today()
    else:
        expenses_date=input("Enter date(format:DD-MM-YYYY): ")

    category=input("Enter category: ")
    amt=float(input("Enter amount: "))
    description=input("Enter description: ")

    with open("expenses.csv","a",newline="") as f:
        writer=csv.writer(f)
        writer.writerow([expenses_id,expenses_date,category,amt,description])
    print("Expenses added Successfully")

def view_expenses():
    with open("expenses.csv","r") as f:
        try:
            reader=csv.reader(f)
            for row in reader:
                print(row)
        except FileNotFoundError:
            print("No data found")

def del_expenses():
  view_expenses()
  records=[]

  with open("expenses.csv","r") as f:
    reader=csv.reader(f)
    for row in reader:
      records.append(row)
  if not records:
    print("The expenses is empty.First enter expenses.")
  else:
    del_id=input("Enter expenses from above to delete: ")
    print(records)
    new_records=[record for record in records if record[0]!=del_id]
    print(new_records)

  if len(new_records)==len(records):
    print("Id not found.")
  else:
    print("Deleted Successfully.")

  with open("expenses.csv","w") as f:
    writer=csv.writer(f)
    for row in new_records:
      writer.writerow(row)

while True:
    print("Enter 1. Add expenses")
    print("2. View expenses")
    print("3. Delete expense")
    print("4. exit")

    choice=int(input("Enter your choice: "))

    if choice==1:
        add_expenses()
    elif choice==2:
        view_expenses()
    elif choice==3:
        del_expenses()
    else:
        print("Thank you...")
        break
```

```python
data={}
def category_summary():
    with open("expenses.csv","r") as f:
        reader=csv.reader(f)
        next(reader)
        for row in reader:
            category=row[2]
            amt=float(row[3])
            data[category]=data.get(category,0)+amt
    print(data)
category_summary()
```

```python
import matplotlib.pyplot as plt

def pie_chart():
    plt.pie(data.values(),labels=data.keys(),autopct="%1.1f%%")
    plt.show()
pie_chart()
```

```python
# Highest Spending Category
max_cat=max(data,key=data.get)
print("Highest Spending category",max_cat)
```

```python
# Monthly expense summary
import csv

def monthly_summary():
    monthly_data={}

    with open("expenses.csv","r") as f:
        reader=csv.reader(f)
        next(reader)
        for row in reader:
            date=row[1]
            amt=float(row[3])
            month=date[3:]

            monthly_data[month]=monthly_data.get(month,0)+amt

    for month in sorted(monthly_data):
        print(month," -> ",monthly_data[month])
monthly_summary()
```
